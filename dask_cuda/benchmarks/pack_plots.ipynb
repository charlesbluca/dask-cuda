```python
import pandas as pd
```

```python
df = pd.read_json("linear_scaling_unpack.json").groupby(["nbytes_str", "columns"], as_index=False).mean()
```

```python
df.pivot(index="columns", columns="nbytes_str", values="took").plot.line(
    title="Unpack wall clock",
    xlabel="# columns",
    ylabel="wall clock (seconds)",
    figsize=(9,9)
)
```
